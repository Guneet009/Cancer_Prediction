```python
numeric_features = [feature for feature in df.columns if df[feature].dtype!='O']
categorical_features = [feature for feature in df.columns if df[feature].dtype=='O'] 
```

```python
print('We have {} numerical features {}'.format(len(numeric_features),numeric_features))
print('We have {} categorical features {}'.format(len(categorical_features),categorical_features))
```

```python
for col in df.columns[::2]:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(df[df['diagnosis'] == 'B'][col], label="Class 0", shade=True)
    sns.kdeplot(df[df['diagnosis'] == 'M'][col], label="Class 1", shade=True)
    plt.title(f'Distribution of {col} by Class')
    plt.legend()
    plt.show()
```

```python
df_input = df.drop(columns=["id","diagnosis"],axis=1)
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(df_input.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
z_scores = np.abs(zscore(df_input))
outliers = np.where(z_scores > 3)
print("Outliers found at indices:", np.unique(outliers[0]))
```

```python
vif_data = pd.DataFrame()
vif_data["Feature"] = df_input.columns
vif_data["VIF"] = [variance_inflation_factor(df_input.values, i) for i in range(df_input.shape[1] )]
print(vif_data)
```
